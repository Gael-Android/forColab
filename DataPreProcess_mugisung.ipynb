```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/tax/data/train.xlsx")
df.head()
# 실행시 18초 정도 걸림
```

```python
df['']
```

```python
new_df = df.drop(columns=['신고번호','신고일자','수입자부호','해외거래처부호','특송업체부호','검사결과코드','우범여부','핵심적발'])
new_df
```

```python
new_df.info()
```

```python
new_df['신고인부호'] = pd.factorize(new_df['신고인부호'])[0]
new_df['수입통관계획코드'] = pd.factorize(new_df['수입통관계획코드'])[0]
new_df['수입신고구분코드'] = pd.factorize(new_df['수입신고구분코드'])[0]
new_df['적출국가코드'] = pd.factorize(new_df['적출국가코드'])[0]

new_df['원산지국가코드'] = pd.factorize(new_df['원산지국가코드'])[0]
new_df['관세율구분코드'] = pd.factorize(new_df['관세율구분코드'])[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

output = min_max_scaler.fit(new_df)
output
```

```python
output = min_max_scaler.transform(new_df)
output = pd.DataFrame(output,columns=new_df.columns)
output['우범여부'] = df['우범여부']
output['핵심적발'] = df['핵심적발']
output
```

```python
output.drop(columns=['우범여부','핵심적발']).to_csv("/content/drive/MyDrive/tax/data/mugisung_X.csv",index=False)
```
